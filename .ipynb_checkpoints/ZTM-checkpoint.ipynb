## Wczytywanie niezbednych modulów do Pythona

```python
%matplotlib inline

import urllib.request # importowanie modulu do otwierania polaczenia z internetem
import json # importowanie modulu do obslugi danych w formacie JSON
import matplotlib.pyplot as plt # importowanie z modulu  "matplotlib" metody "pyplot" oznaczonej jako "plt" - tworzenie grafiki
import logging # import modulu  logging - zarzadzanie praca innych modulow - nie wiem o co w tym chodzi :P
import geotiler # import modulu geotiler - rysowanie map
```

## Wczytywanie danych z internetu

```python
'''
Ponizszy objekt "adres" to zapytanie do bazy danych z warszawskimi tramwajami.
Możecie skopiowac to zapytanie i wkleic je do przegladarki jako zwykly adres.
W odpowiedzi wyswietli Wam sie baza danych - nie jest duza.
'''
adres='https://api.um.warszawa.pl/api/action/wsstore_get/?id=c7238cfe-8b1f-4c38-bb4a-de386db7e776&apikey=tutaj_wpisz_swoj_klucz'
polaczenie = urllib.request.urlopen(adres) # otwieranie polaczenia z baza
tekst = polaczenie.read() # wczytywanie do Pythona do obiektu "tekst" danych ze strony - dane wczytuja sie jako typ 'bytes'
tekst = tekst.decode() # zeby dane dalo sie parsowac jako plik JSON trzeba je rekodowac do typu 'str' czyli do zwyklego tekstu
print( type(tekst)) # podglad typu obiektu "tekst"
```

## Parsowanie bazy pobranej z Internetu 

```python
'''
Parsowanie danych - przeksztalcanie danych ze zwyklego tekstu na baze danych z zapisie JSON
'''
ztm = json.loads(tekst) # parsowanie danych i tworzenie obiektu "ztm" - dane sa teraz typu 'dict'
ztm_liczba = len(ztm['result']) # sprawdzanie ile tramwajow znajduje sie w bazie danych
print( ztm['result'][0]) # podglad danych dla pierwszego tramwaju w zbiorze (Python zaczyna numeracje od 0)
```

## Dokładny czas i data pobrania danych

```python
print( "Pozycje tramwajów warszawskich zgodnie ze stanem na ", ztm['result'][0]['Time'])
```

## Przepisywanie danych z bazy do obiektu typu lista

```python
'''
Ponizej znajduje sie petla ktora do obietky "lista" typu 'list' 
dodaje jako kolejne elementy listy wspolrzedne tramwajow.
Pentla zaczyna sie od 0 bo dla Pythona pierwszy element obiektu zawsze ma numer zero.
Petla przechodzi przez poszczgolne obiekty bazy (indeks i), 
a następnie wyciąga z nich dane zapisane jako 'Lat' oraz 'Lon' (Latitude, Longitude)
'''
lista = [] # tworzenie pustej listy do ktorej beda wpisywane dane o polozeniu tramwajow
for i in range(0, ztm_liczba-1) : # petla
    if ztm['result'][i]['Lon']>0: # nie wiem dlaczego, ale niektore tramwaje maja lokalizacje (0,0) - wybieram pozostale
        lista.append( ( ztm['result'][i]['Lon'], ztm['result'][i]['Lat'])    )
print( lista[0]) # podglad pierwszego obiektu na liscie
```

## Tworzenie mapy

```python
tramwaje=tuple(lista) # zamienianie listy w obiekt typu 'tuple' - to jest potrzebne do naniesienia punktow na mape

bbox = 20.89647, 52.14750, 21.12100, 52.32630 # wybieranie obszaru na mapie
# bbox = 20.90500, 52.14750, 21.12100, 52.32630


fig = plt.figure(figsize=(30, 25)) # wyznaczanie wielkosci obrazu
ax = plt.subplot(111) # tworzenie pola wykresu

mm = geotiler.Map(extent=bbox, zoom=15) # pobieranie mapy

img = geotiler.render_map(mm) # renderowanie mapy na obraz
ax.imshow(img) # wklejanie obrazu do wykresu

x, y = zip(*(mm.rev_geocode(p) for p in tramwaje)) # dodawania lokalizacji tramwajow do mapy
ax.scatter(x, y, c='red', edgecolor='none', s=100, alpha=0.9)
```
